```python
import pandas as pd
```

```python
tw=pd.read_csv(r'Tweetss.csv',sep='~')
tw
```

```python
tw['hashtags']=tw.hashtags.str.replace("[", '')
tw['hashtags']=tw.hashtags.str.replace("]", '')
tw['hashtags']=tw.hashtags.str.replace("#", '')
tw['hashtags']=tw.hashtags.str.replace("'", '')
```

```python
tw
```

```python
tw['hashtags'].str.count(",").sort_values(ascending=False)
```

```python
df_h=tw['hashtags'].str.split(",",expand=True)
df_h
```

```python
from itertools import product,combinations  
```

```python
a=[]
for i in range(len(df_h)):
    if len(df_h.iloc[i].dropna())>=3:
        # print(df_h.iloc[i].dropna())

        a.append(list(combinations(df_h.iloc[i].dropna(),2)))
        
    
```

```python
source=[]
target=[]
for item in a:
    for j in range(3):
        source.append(item[j][0])
        target.append(item[j][1])
    
```

```python
edges=pd.DataFrame()
```

```python
edges['Source']=source
edges['Target']=target
```

```python
edges.to_csv('edges.csv',index=False)
```
